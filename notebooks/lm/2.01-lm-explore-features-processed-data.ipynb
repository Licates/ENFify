```python
import json

import matplotlib.pyplot as plt
import numpy as np

from enfify import INTERIM_DATA_DIR, PROCESSED_DATA_DIR
```

```python
dsets = sorted([f for f in PROCESSED_DATA_DIR.iterdir() if f.is_dir()])
sample_files = [sorted(dset.glob("*.npy"))[:2] for dset in dsets]
```

```python
# plot samples
# for (auth_file, tamp_file), cut_info_path in zip(sample_files, cut_info_paths):
sample_files = sorted(PROCESSED_DATA_DIR.glob("Carioca2/*.npy"))
sample_files = [(sample_files[i], sample_files[i+1]) for i in range(0, len(sample_files), 2)]

for auth_file, tamp_file in sample_files[:10]:
    auth_sig = np.load(auth_file)
    tamp_sig = np.load(tamp_file)
    dset = auth_file.parent.name
    plt.figure(figsize=(20,5))
    plt.plot(auth_sig, label=auth_file.stem)
    plt.plot(tamp_sig, label=tamp_file.stem)
    plt.title(dset)
    plt.xlabel("Frames")
    plt.ylabel("Frequency (Hz)")
    plt.legend()
    plt.show()
```
